```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

**TASK # 1**

```python
img = cv.imread("/content/drive/MyDrive/mountain.jpg")
cv2_imshow(img)
img.shape
```

```python
grayImg = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(grayImg)
```

```python
hist = cv.calcHist([grayImg], [0], None, [256], [0, 256])
plt.figure()
plt.title("GrayScale Histogram")
plt.xlabel("Gray Levels")
plt.ylabel("No. of pixels")
plt.plot(hist, color="black")
plt.xlim([0, 256])
plt.show()
```

```python
equImg = cv.equalizeHist(grayImg)
cv2_imshow(equImg)
```

```python
equHist = cv.calcHist([equImg], [0], None, [256], [0, 256])

plt.figure()
plt.title("Grayscale Histogram of Equalized Image")
plt.xlabel("Gray Levels")
plt.ylabel("No. of Pixels")
plt.plot(equHist, color="black")
plt.xlim([0, 256])
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# original grayscale image
plt.subplot(2, 2, 1)
plt.imshow(grayImg, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis("off")

# equalized grayscale image
plt.subplot(2, 2, 2)
plt.imshow(equImg, cmap='gray')
plt.title("Equalized Grayscale Image")
plt.axis("off")

# Grayscale Image Histogram
plt.subplot(2, 2, 3)
plt.plot(hist, color="black")
plt.title("GrayScale Image Histogram")
plt.xlabel("Gray Levels")
plt.ylabel("No. of pixels")
plt.xlim([0, 256])

# Grayscale Image Equalized Histogram
plt.subplot(2, 2, 4)
plt.plot(equHist, color="red")
plt.title("GrayScale Image Equalized Histogram")
plt.xlabel("Gray Levels")
plt.ylabel("No. of pixels")
plt.xlim([0, 256])
```

```python
img = cv.imread("/content/drive/MyDrive/view.jpg")
cv2_imshow(img)
img.shape
```

```python
grayImg = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(grayImg)
```

```python
binaryImg = cv.threshold(grayImg, 127, 255, cv.THRESH_BINARY)
cv2_imshow(binaryImg[1])
```

```python
plt.figure(figsize=(12, 6))

# original grayscale image
plt.subplot(1, 2, 1)
plt.imshow(grayImg, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis("off")

# binary image
plt.subplot(1, 2, 2)
plt.imshow(binaryImg[1], cmap="gray")
plt.title("Binary Image")
plt.axis("off")
```
